```python
import pandas as pd 
import matplotlib
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
```

```python
def sinplot(flip = 1):
    x = np.linspace(0, 14, 100)
    for i in range(1, 5):
        plt.plot(x, np.sin(x + i * .5) * (7 - i) * flip)
```

```python
sns.set_style("ticks")
sns.set_palette("pastel")
sinplot()
plt.show()
```

```python
tips_df = pd.read_csv('tips.csv')
tips_df["tip_percentage"] = tips_df["tip"] / tips_df["total_bill"]
pivot = tips_df.pivot_table(
    index=["day"],
    columns=["size"],
    values="tip_percentage",
    aggfunc=np.average)
sns.heatmap(pivot)
sns.heatmap(pivot, cmap="RdBu")
sns.heatmap(pivot, cmap="Spectral")
```
